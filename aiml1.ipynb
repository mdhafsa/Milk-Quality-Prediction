```python
x=d.iloc[:,0:7]
y=d.iloc[:,7:8]
print(x)
print(y)
```

```python
from sklearn.model_selection import train_test_split
print(x.shape)
x_train,x_test,y_train,y_test=train_test_split(x,y,random_state=True,test_size=0.20)
```

```python
from sklearn.linear_model import LogisticRegression
lr=LogisticRegression()
mm=lr.fit(x_train,y_train)
```

```python
yp=mm.predict(x_test)
print(yp)
```

```python
from sklearn.metrics import accuracy_score
print(accuracy_score(yp,y_test))
```

```python
from sklearn import metrics
```

```python
confusion_matrix=metrics.confusion_matrix(y_test,yp)
```

```python
cm_display=metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix,display_labels=[3,2,1])
```

```python
import matplotlib.pyplot as plt
```

```python
cm_display.plot()
plt.show()
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(criterion='entropy',random_state=0)
mm=classifier.fit(x_train,y_train)
```

```python
yp=mm.predict(x_test)
```

```python
print(yp)
```
